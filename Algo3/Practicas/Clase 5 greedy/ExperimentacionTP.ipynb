# Experimentaci贸n del TP

````python
#Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def nlogn(x):
    return x * np.log(x);
fig = plt.figure()
````

### Instancias random

Primero podemos ver la correlaci贸n de nuestras medidas con la cota te贸rica que estimamos. Para esto usamos 

```
sns.lmplot
```

que hace una regresi贸n lineal sobre las dos columnas que le pasamos.

````python
df_random = pd.read_csv("runtime_tp_random.csv")
df_random['cota'] = nlogn(df_random['n'])

ax1 = sns.lmplot(x='time', y='cota', data=df_random)
plt.xlabel("tiempo (s)")
plt.ylabel("cota teorica")
````

Este gr谩fico tiene el problema de que los n que consideramos crecen exponencialmente. Podemos hacer el plot con escala logaritmica para que se vea mejor.

````python
df_random['log_time'] = np.log(df_random['time'])
df_random['log_cota'] = np.log(df_random['cota'])
ax1 = sns.lmplot(x='log_time', y='log_cota', data=df_random)
plt.xlabel("log_tiempo (s)")
plt.ylabel("log_cota")
````

Incluso podemos calcular el coeficiente de Pearson, que es una medida de dependencia lineal entre variables. Mientras m谩s cerca est茅 de 1, m谩s confianza tenemos de que nuestras variables est谩n linealmente correlacionadas.

````python
r = np.corrcoef(df_random['log_time'], df_random['log_cota'])[0, 1];
print("El coeficiente de correlaci贸n de Pearson es: r="+str(r))
````
